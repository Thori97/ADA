```python
import torch, torchvision
from torchvision import datasets, transforms
import numpy as np
import matplotlib.pyplot as plt
```

```python
device = torch.device('cpu')
#device = torch.device('gpu')
```

```python
train_dataset = torchvision.datasets.MNIST(root='./data', 
                                        train=True,
                                        download=True,
                                        transform=transforms.ToTensor())
```

```python
validation_dataset = torchvision.datasets.MNIST(root='./data', 
                                        train=False,
                                        download=True,
                                        transform=transforms.ToTensor())
```

```python
batch_size = 32
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_size = 5
validation_loader = torch.utils.data.DataLoader(dataset=validation_dataset, batch_size=test_size, shuffle=True)
```

```python
import torch.nn as nn
import torch.nn.functional as F
class AutoEncoder(nn.Module):
    def __init__(self):
        med = 80
        super(AutoEncoder, self).__init__()
        self.fc1 = nn.Linear(28*28, med)
        self.fc2 = nn.Linear(med, 28*28)
    def forward(self, x):
        x = x.view(-1, 28*28)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        x = torch.sigmoid(x)
        return x

model = AutoEncoder().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)
criterion = nn.MSELoss()

print(model)
```

```python
def train(epoch, log_interval=200):
    # Set model to training mode
    model.train()
    
    # Loop over each batch from the training set
    for batch_idx, (data, target) in enumerate(train_loader):

        # Zero gradient buffers
        optimizer.zero_grad() 
        
        # Pass data through the network
        output = model(data)

        # Calculate loss
        data = data.reshape(batch_size, 28*28)
        #print(output.shape, data.shape)
        loss = criterion(output, data)

        # Backpropagate
        loss.backward()
        
        # Update weights
        optimizer.step()
        
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.data.item()))
```

```python
train(50)
```

```python
def add_noise(im):
    p = np.random.rand(28, 28)
    im = im.detach().numpy()
    im[p<0.05] = 1
    return im
def validate():
    model.eval()
    size = 3
    fig = plt.figure(figsize=(size*test_size, size))
    num = 0
    for data, target in validation_loader:
        data = data.to(device)
        for i in range(test_size):
            ax = fig.add_subplot(2, 5, i+1)
            ax.imshow(add_noise(data[i][0]))
            output = model(data[i][0])
            ax = fig.add_subplot(2, 5, i+6)
            output = output.detach().numpy()
            ax.imshow(output.reshape(28, 28))
        break
    plt.savefig("autoencoder.pdf")
    plt.show()
```

```python
validate()
```
