```python
import torch, torchvision
from torchvision import datasets, transforms
import numpy as np
import matplotlib.pyplot as plt
```

```python
device = torch.device('cpu')
```

```python
train_dataset = torchvision.datasets.MNIST(root='./data', 
                                        train=True,
                                        download=True,
                                        transform=transforms.ToTensor())
validation_dataset = torchvision.datasets.MNIST(root='./data', 
                                        train=False,
                                        download=True,
                                        transform=transforms.ToTensor())
```

```python
batch_size = 32
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
validation_loader = torch.utils.data.DataLoader(dataset=validation_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch.nn as nn
import torch.nn.functional as F
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(28*28, 50)
        self.fc2 = nn.Linear(50, 10)
    def forward(self, x):
        x = x.view(-1, 28*28)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
        #return torch.sigmoid(x)

model = Net().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)
#optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)
criterion = nn.CrossEntropyLoss()

print(model)
```

```python
def train(epoch, log_interval=200):
    # Set model to training mode
    model.train()
    l = 0
    # Loop over each batch from the training set
    for batch_idx, (data, target) in enumerate(train_loader):

        # Zero gradient buffers
        optimizer.zero_grad() 
        
        # Pass data through the network
        output = model(data)

        # Calculate loss
        loss = criterion(output, target)

        # Backpropagate
        loss.backward()
        
        # Update weights
        optimizer.step()
        
        l = loss.data.item()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.data.item()))
    lossv_train.append(l)
def validate(loss_vector, accuracy_vector):
    model.eval()
    val_loss, correct = 0, 0
    for data, target in validation_loader:
        data = data.to(device)
        target = target.to(device)
        output = model(data)
        val_loss += criterion(output, target).data.item()
        pred = output.data.max(1)[1]
        correct += pred.eq(target.data).cpu().sum()
    
    val_loss /= len(validation_loader)
    loss_vector.append(val_loss)
    
    accuracy = 100. * correct.to(torch.float32) / len(validation_loader.dataset)
    accuracy_vector.append(accuracy)
    
    print('\nValidation set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        val_loss, correct, len(validation_loader.dataset), accuracy))
```

# 早期打ち切り

```python
epochs = 20

lossv_train = []
lossv_valid, accv_valid = [], []
for epoch in range(1, epochs + 1):
    train(epoch)
    validate(lossv_valid, accv_valid)
```

```python
plt.figure(figsize=(5, 3))
plt.plot(np.arange(1, epochs+1), lossv_train, label="train")
plt.plot(np.arange(1, epochs+1), lossv_valid, label="valid")
plt.title('early stop loss')
m = np.argmin(lossv_valid)
plt.scatter(m+1, lossv_valid[m], facecolor='None', edgecolors='red', s = 100)
plt.legend()
plt.savefig("bigreport/earlystop.pdf")
```

# ラベル平滑化

```python
def linear_combination(x, y, epsilon):
    return (1 - epsilon) * x + epsilon * y

def reduce_loss(loss, reduction='mean'):
    return loss.mean() if reduction == 'mean' else loss.sum() if reduction == 'sum' else loss

class CustomLoss(nn.Module):
    def __init__(self, alpha=0.1):
        super().__init__()
        self.alpha = alpha

    def forward(self, outputs, targets):
        n = outputs.size()[-1]
        log_outputs = F.log_softmax(outputs, dim=-1)
        loss = reduce_loss(-log_outputs.sum(dim=-1), 'mean')
        nll = F.nll_loss(log_outputs, targets, reduction='mean')
        return linear_combination(nll, loss/n, self.alpha)
```

```python
import torch.nn as nn
import torch.nn.functional as F
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(28*28, 50)
        self.fc2 = nn.Linear(50, 10)
    def forward(self, x):
        x = x.view(-1, 28*28)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
        #return torch.sigmoid(x)

model = Net().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)
criterion = CustomLoss(0.2)

print(model)
```

# ドロップアウト

```python
import torch.nn as nn
import torch.nn.functional as F
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(28*28, 50)
        self.fc2 = nn.Linear(50, 10)
    def forward(self, x):
        x = x.view(-1, 28*28)
        x = F.relu(self.fc1(x))
        x = F.dropout(x, 0.3)
        x = self.fc2(x)
        return x

model = Net().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)
criterion = nn.CrossEntropyLoss()

print(model)
```

```python
epochs = 10

lossv_train = []
lossv_valid, accv_valid = [], []
for epoch in range(1, epochs + 1):
    train(epoch)
    validate(lossv_valid, accv_valid)
```

```python
plt.figure(figsize=(5, 3))
plt.plot(np.arange(1, epochs+1), lossv_train, label="train")
plt.plot(np.arange(1, epochs+1), lossv_valid, label="valid")
plt.title('dropout loss')
plt.legend()
plt.savefig("bigreport/dropout.pdf")
```
