線形モデル

$$f_\theta (x) = \theta ^\top x + \theta_0$$

に対して, クラス比重み付き最小二乗法を実装せよ

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  データ生成
n = 100
n1 = n//10*9
n2 = n//10
x = np.random.randn(n, 2)
x[:n1, 0] -= 2
x[n1:, 0] += 2
x[:, 1] *= 2
y = np.ones((n,))
y[n1:] *= -1

X = np.random.randn(n, 2)
X[:n2, 0] -= 2
X[n2:, 0] += 2
X[:, 1] *= 2
```

```python
#  クラス比推定
def dist(x, y):
    return np.linalg.norm(x-y, ord=2)
App = sum([sum([dist(x[i, :], X[j, :]) for j in range(100) if y[j]==1 ]) for i in range(n) if y[i]==1 ])/n1/n1
Apm = sum([sum([dist(x[i, :], X[j, :]) for j in range(100) if y[j]==-1 ]) for i in range(n) if y[i]==1 ])/n1/n2
Amp = sum([sum([dist(x[i, :], X[j, :]) for j in range(100) if y[j]==1 ]) for i in range(n) if y[i]==-1 ])/n2/n1
Amm = sum([sum([dist(x[i, :], X[j, :]) for j in range(100) if y[j]==-1 ]) for i in range(n) if y[i]==-1 ])/n2/n2
bp = sum([sum([dist(X[i, :], x[j, :]) for j in range(100) if y[j]==1 ]) for i in range(n)])/n/n1
bm = sum([sum([dist(X[i, :], x[j, :]) for j in range(100) if y[j]==-1 ]) for i in range(n)])/n/n2

pi_tilde = (Apm-Amm-bp+bm)/(2*Apm-App-Amm)
pi = min(1, max(0, pi_tilde))
pi
```

```python
# 見苦しいですがweightedのパラメータ計算　多分合ってる
# やってから思ったのですが適当にnumpyとかで最適化すると楽そう
n1l, n2l = [i for i in range(0, n1)], [i for i in range(n1, n)]
a = 0.1/(1-pi)
b = 0.9/pi
A = a*sum([x[i, 0]**2 for i in n1l])+b*sum([x[i, 0]**2 for i in n2l])
B = a*sum([x[i, 1]**2 for i in n1l])+b*sum([x[i, 1]**2 for i in n2l])
C = a*n1+b*n2
D = 2*a*sum([x[i, 0]*x[i, 1] for i in n1l])+2*b*sum([x[i, 0]*x[i, 1] for i in n2l])
E = 2*a*sum([x[i, 1] for i in n1l])+2*b*sum([x[i, 1] for i in n2l])
F = 2*a*sum([x[i, 0] for i in n1l])+2*b*sum([x[i, 0] for i in n2l])
G = -2*a*n1+2*b*n2
H = -2*a*sum([x[i, 0] for i in n1l])+2*b*sum([x[i, 0] for i in n2l])
I = -2*a*sum([x[i, 1] for i in n1l])+2*b*sum([x[i, 1] for i in n2l])
```

```python
M = np.array([[2*A, D, F], [D, 2*B, E], [F, E, 2*C]])
thetaa, thetab, thetac = np.linalg.inv(M).T@np.array([-H, -I, -G])
```

```python
#  unweightedのパラメータ計算
psi = np.ones((n, 3))
for i in range(n):
    psi[i, 0], psi[i, 1] = x[i, 0], x[i, 1] 
theta1, theta2, theta3 = np.linalg.inv(psi.T@psi)@psi.T@y
```

```python
#  プロット(訓練データ)
plt.figure(figsize=(5,5))
xx = np.linspace(-5, 5)
yy1 = (thetaa*xx+thetac)/(-1*thetab)
yy2 = (theta1*xx+theta3)/(-1*theta2)
plt.plot(xx, yy1, label = 'weighted')
plt.plot(xx, yy2, label = 'unweighted')
plt.legend()
plt.scatter(x[:n1, 0], x[:n1, 1], facecolor='None', edgecolor='blue')
plt.scatter(x[n1:, 0], x[n1:, 1], facecolor='None', edgecolor='red')
plt.xlim(-5, 5)
plt.ylim(-10, 10)
plt.title('train data')
plt.show()
```

```python
#  プロット(テストデータ)
plt.figure(figsize=(5,5))
xx = np.linspace(-5, 5)
yy1 = (thetaa*xx+thetac)/(-1*thetab)
yy2 = (theta1*xx+theta3)/(-1*theta2)
plt.plot(xx, yy1, label = 'weighted')
plt.plot(xx, yy2, label = 'unweighted')
plt.legend()
plt.scatter(X[:n2, 0], X[:n2, 1], facecolor='None', edgecolor='blue')
plt.scatter(X[n2:, 0], X[n2:, 1], facecolor='None', edgecolor='red')
plt.xlim(-5, 5)
plt.ylim(-10, 10)
plt.title('test data')
plt.show()
```

↑weightedはテストデータをよく分離している、unweightedではうまく分類できていない事がわかる
