```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)  # set the random seed for reproducibility
```

```python
def generate_sample(xmin, xmax, sample_size):
    x = np.linspace(start=xmin, stop=xmax, num=sample_size)
    pix = np.pi * x
    target = np.sin(pix) / pix + 0.1 * x
    noise = 0.05 * np.random.normal(loc=0., scale=1., size=sample_size)
    return x, target + noise


def calc_design_matrix(x, c, h):
    return np.exp(-(x[None] - c[:, None]) ** 2 / (2 * h ** 2))
```

## $h$と$l$をいくつか用意してそれぞれに対してlossを計算する

一つ抜き交差確認を用い、宿題2で証明した式でlossを計算した

```python
# create sample
sample_size = 50
xmin, xmax = -3, 3
x, y = generate_sample(xmin=xmin, xmax=xmax, sample_size=sample_size)


hs = np.logspace(-2, 0.5, 20)

ls = np.logspace(-5, -1, 50)
```

```python
min_h = 1e10
min_l = 1e10
min_loss = 1e10
for h in hs:
    for l in ls:
        k = calc_design_matrix(x, x, h)
        I = np.identity(len(k))
        H = I - k@np.linalg.inv(k.T@k + l * I)@k.T
        H_hat = H * I
        #print(H)
        #print(H_hat)
        loss = np.linalg.norm(np.linalg.inv(H_hat)@H@y)/sample_size
        print(f"h={h:.03f}\tl={l:.05f}\tloss={loss:.06f}")
        if loss < min_loss:
            min_loss = loss
            min_h = h
            min_l = l
```

```python
print(f"h, l = ({min_h:.03f}, {min_l:.05f}) \nis the best")
```

```python
k = calc_design_matrix(x, x, min_h)
theta = np.linalg.solve(
k.T.dot(k) + min_l * np.identity(len(k)),
k.T.dot(y[:, None]))

# create data to visualize the prediction
X = np.linspace(start=xmin, stop=xmax, num=5000)
K = calc_design_matrix(x, X, min_h)
prediction = K.dot(theta)

# visualization
plt.clf()
plt.scatter(x, y, c='green', marker='o', label="data")
plt.plot(X, np.sin(np.pi*X)/(np.pi*X) + 0.1*X, c = "red", linewidth = 3, label = "true")
plt.plot(X, prediction, c="blue", linewidth = 3, label ="estimate")
plt.legend()
plt.show()
```
